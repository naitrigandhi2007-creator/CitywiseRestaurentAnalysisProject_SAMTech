# **Introduction**

This project analyzes restaurant data from the Zomato dataset to identify city-wise restaurant distribution, customer ratings, pricing trends, cuisine preferences, and restaurant popularity.

The objective is to generate meaningful business insights through Exploratory Data Analysis (EDA) and data visualization.



# **## Dataset Overview**

The dataset contains information about restaurants including:

- Restaurant Name
- City
- Cuisine
- Average Rating
- Votes
- Prices
- Restaurant Popularity
- Best Seller Status

Total Records: 123,657

Total Features: 26

# ***## Data Cleaning***


### Observation

- No missing values were found.
- 22,127 duplicate rows were removed.

```python
import pandas as pd
import pandas as pd

df = pd.read_csv("enhanced_zomato_dataset_clean.csv")

print(df.head())
```

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nInfo:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

# **# Exploratory Data Analysis (EDA)**

```python
df.drop_duplicates(inplace=True)

print("New Shape:", df.shape)
print("Duplicate Rows:", df.duplicated().sum())
```

```python
city_count = df["City"].value_counts()

print(city_count)
```

# **### 1. Cities with Highest Number of Restaurants**

```python
top10_cities = df["City"].value_counts().head(10)

print(top10_cities)
```

```python
import matplotlib.pyplot as plt

top10_cities.plot(kind="bar", figsize=(10,6))

plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)

plt.show()
```

# **### 2. Average Ratings by City**

```python
avg_rating_city = df.groupby("City")["Average_Rating"].mean().sort_values(ascending=False)

print(avg_rating_city)
```

```python
top10_rating = avg_rating_city.head(10)

print(top10_rating)
```

```python
import matplotlib.pyplot as plt

top10_rating.plot(kind="bar", figsize=(10,6))

plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)

plt.show()
```

# **### 3. Average Restaurant prices by City**

```python
avg_price = df.groupby("City")["Prices"].mean().sort_values(ascending=False)

print(avg_price.head(10))
```

```python
top10_price = avg_price.head(10)

top10_price.plot(kind="bar", figsize=(10,6))

plt.title("Top 10 Cities by Average Restaurant Price")
plt.xlabel("City")
plt.ylabel("Average Price")

plt.xticks(rotation=45)
plt.show()
```

# ***## 4. Popular Cuisines***

```python
cuisine_count = df["Cuisine"].value_counts()

print(cuisine_count.head(10))
```

```python
top10_cuisine = cuisine_count.head(10)

top10_cuisine.plot(kind="bar", figsize=(12,6))

plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)

plt.show()
```

# **##5. Restaurant Popularity**

```python
top_restaurants = df.groupby("Restaurant_Name")["Votes"].sum().sort_values(ascending=False)

print(top_restaurants.head(10))
```

```python
top10_restaurants = top_restaurants.head(10)

top10_restaurants.plot(kind="bar", figsize=(12,6))

plt.title("Top 10 Restaurants by Votes")
plt.xlabel("Restaurant")
plt.ylabel("Total Votes")
plt.xticks(rotation=45)

plt.show()
```

# **## 6. Histogram**

```python
plt.figure(figsize=(8,5))

plt.hist(df["Average_Rating"], bins=10)

plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Number of Restaurants")

plt.show()
```

# **## 7. Prices v/s Average Ratings**

```python
plt.figure(figsize=(8,6))

plt.scatter(df["Prices"], df["Average_Rating"])

plt.title("Price vs Average Rating")
plt.xlabel("Price")
plt.ylabel("Average Rating")

plt.show()
```

# **## 8. Correlation Analysis**

```python
correlation = df[[
    "Average_Rating",
    "Prices",
    "Votes",
    "Dining_Votes",
    "Delivery_Votes",
    "Restaurant_Popularity"
]].corr()

print(correlation)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.imshow(correlation, cmap="coolwarm", interpolation="nearest")
plt.colorbar()

plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation.columns)), correlation.columns)

plt.title("Correlation Matrix")

plt.show()
```

```python
high_rating = df["Is_Highly_Rated"].value_counts()

print(high_rating)
```

# **## 9. Scatter Plot**

```python
high_rating.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Highly Rated vs Normal Restaurants")
plt.ylabel("")
plt.show()
```

```python
expensive = df["Is_Expensive"].value_counts()

print(expensive)
```

```python
expensive.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Expensive vs Affordable Restaurants")
plt.ylabel("")
plt.show()
```

```python
popular = df.groupby("Restaurant_Name")["Restaurant_Popularity"].mean().sort_values(ascending=False)

print(popular.head(10))
```

```python
popular.head(10).plot(kind="bar", figsize=(12,6))

plt.title("Top 10 Most Popular Restaurants")

plt.xlabel("Restaurant")

plt.ylabel("Popularity Score")

plt.xticks(rotation=45)

plt.show()
```

```python
price_cuisine = df.groupby("Cuisine")["Prices"].mean().sort_values(ascending=False)

print(price_cuisine.head(10))
```

```python
price_cuisine.head(10).plot(kind="bar", figsize=(12,6))

plt.title("Most Expensive Cuisines")

plt.xlabel("Cuisine")

plt.ylabel("Average Price")

plt.xticks(rotation=45)

plt.show()
```

```python
rating_cuisine = df.groupby("Cuisine")["Average_Rating"].mean().sort_values(ascending=False)

print(rating_cuisine.head(10))
```

```python
rating_cuisine.head(10).plot(kind="bar", figsize=(12,6))

plt.title("Highest Rated Cuisines")

plt.xlabel("Cuisine")

plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.show()
```
